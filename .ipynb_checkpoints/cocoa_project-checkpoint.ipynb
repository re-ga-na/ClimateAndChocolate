## Data Cleaning

```python
# Get the column names
flavors_cacao.columns
```

```python
# Rename columns
flavors_cacao = flavors_cacao.rename(columns={"Company \n(Maker-if known)": "Company",
                                                      "Specific Bean Origin\nor Bar Name": "Specific Bean Origin",
                                                      "Review\nDate": "Review Date",
                                                      "Cocoa\nPercent": "Cocoa Percent",
                                                      "Company\nLocation": "Company Location",
                                                      "Bean\nType": "Bean Type",
                                                      "Broad Bean\nOrigin": "Broad Bean Origin"
                             })
flavors_cacao.columns = [c.replace('\n', ' ') for c in flavors_cacao.columns]
flavors_cacao.head()
```

```python
# Convert object to float, replace % symbol
flavors_cacao['Cocoa Percent'] = flavors_cacao['Cocoa Percent'].str.replace(r'%', r' ').astype('float')
# flavors_cacao['Cocoa Percent']=(flavors_cacao['Cocoa Percent']).str.replace('%', ' ')
# flavors_cacao['Cocoa Percent']=(flavors_cacao['Cocoa Percent']).astype(float)
# flavors_cacao['Cocoa Percent'].value_counts()
```

```python
# Fill space with NaN
flavors_cacao = flavors_cacao.replace(r'^\s*$', np.nan, regex=True)
flavors_cacao.head(10)
```

```python
# Drop NaN values from the Broad Bean Origin column only
dropna_flavors_cacao = flavors_cacao.dropna(subset=['Broad Bean Origin'])
dropna_flavors_cacao.head(10)
```

1795 chocolate bars of 60 countries were rated from 2006 to 2017. 
Most of chocolate bars (392, 341, 303) were rated between 3.0 to 3.5. 
Most of the chocolate has 70 to 75 percent of cocoa. 
Most of the rated chocolate is grown in South America (with exception of Madagascar).
The number one grower is Venezuela. 
The most rated bean is Trinitario, followed by Forastero and Criollo

## James - GDP

```python
count_of_rating_by_origin = flavors_cacao.groupby(["Broad Bean Origin"])
counts_rating_by_origin = pd.DataFrame(count_of_rating_by_origin["Rating"].count())
counts_rating_by_origin.reset_index(inplace=True)
#pick the top5 most reviews by bean origin 
counts_rating_by_origin = counts_rating_by_origin.rename(columns={"Rating":"counts_of_rating"})
rank_counts_rating_origin = counts_rating_by_origin.sort_values(by=['counts_of_rating'],ascending=False)
rank_counts_rating_origin.set_index("Broad Bean Origin", inplace=True)
most_counts_by_origin = rank_counts_rating_origin.iloc[0:10]
most_counts_by_origin.dropna()
```

```python
bean_origin_places = "Resources/GDP_bean_production_current$_v2.csv"
bean_origin_gdps = pd.read_csv(bean_origin_places)
bean_origin_gdps = bean_origin_gdps.rename(columns={"Broad Bean\nOrigin": "Broad Bean Origin"})
bean_origin_gdps = bean_origin_gdps.drop_duplicates()
bean_origin_gdps.set_index("Broad Bean Origin", inplace=True)
bean_origin_gdps
```

```python
top_origins = bean_origin_gdps.loc[["Venezuela","Ecuador","Peru","Madagascar","Trinidad","Vietnam","Brazil","Ghana","Belize","Jamaica"]]
top_origins = top_origins.drop_duplicates()
top_origins = top_origins.drop(columns=['Income group', 'Region'])
top_origins = top_origins.apply(pd.to_numeric)
top_origins['avg_gdp'] = top_origins.mean(axis=1)
top_origins_avg_gdp = pd.DataFrame(top_origins['avg_gdp'])
top_origins_avg_gdp
```

```python
gdp_vs_most_counts = most_counts_by_origin.join(top_origins_avg_gdp,how = "outer")
gdp_vs_most_counts.dropna()
```

```python
plt.scatter(gdp_vs_most_counts['counts_of_rating'], gdp_vs_most_counts['avg_gdp'])
plt.title(f'counts of ratings vs. GDPs')
plt.xlabel('counts_of_rating')
plt.ylabel('average GDPs')
plt.grid(True)
plt.savefig("Images/Counts_of_ratings_vs_bean_origin_GDPs.png")
plt.show()
```

```python
rating_by_origin = flavors_cacao.groupby(["Broad Bean Origin"])
mean_rating_by_origin = pd.DataFrame(rating_by_origin["Rating"].mean())
mean_rating_by_origin.reset_index(inplace=True)
mean_rating_by_origin
rank_mean_rating_origin = mean_rating_by_origin.sort_values(by=['Rating'],ascending=False)
rank_mean_rating_origin.set_index("Broad Bean Origin", inplace=True)
top_rating_by_origin = rank_mean_rating_origin.loc[["Venezuela","Ecuador","Peru","Madagascar","Trinidad","Vietnam","Brazil","Ghana","Belize","Jamaica"]]
top_rating_by_origin
```

```python
plt.scatter(gdp_vs_top_ratings['Rating'], gdp_vs_top_ratings['avg_gdp'])
plt.title(f'ratings vs. GDPs')
plt.xlabel('ratings')
plt.ylabel('average GDPs')
plt.grid(True)
plt.savefig("Images/Ratings_vs_bean_origin_GDPs.png")
plt.show()
```

## Regana - Climate

```python
#feeding in csv path with global land temp data from all countries measured monthly since 1743.
csv_path = "Resources/GlobalLandTemperaturesByCountry.csv"
country_temps = pd.read_csv(csv_path)
country_temps.head()
```

```python
#use count to check for missing entries
country_temps.count()
```

```python
#check data types
country_temps.dtypes
```

```python
#find column names
country_temps = pd.DataFrame(country_temps) 
country_temps.columns
```

```python
#renaming columns
country_temps = country_temps.rename(columns={"dt": "Reading Date", "AverageTemperature":"Temperature", "AverageTemperatureUncertainty": "Average Temperature Uncertainty", "Country": "Country"})

country_temps.head(10)
```

```python
#Removes lines without temp. readings

dropna_country_temps = country_temps.dropna(subset=['Temperature'])
dropna_country_temps.head(10)
```

```python
#all the columns have the same amount of data!

dropna_country_temps.count()
```

```python
#listing out names of countries w temp recordings
dropna_country_temps.Country.unique()
```

```python
#indexing data by country name
df = dropna_country_temps.set_index("Country")
df
```

```python
#removes dashes from dates and converts them to int
dasheddates = df["Reading Date"]

df["Reading Date"] = df["Reading Date"].replace('-', '', regex=True)


df["Reading Date"] = df["Reading Date"].astype(int)

df.dtypes
```

```python
# #isolates data from 20 cacao bean growing nations that produced beans for the best rated chocolate bars, as determined in the previous jupyter notebook
df = df.loc[['Belize', 'Brazil', 'Ecuador', 'Ghana', 'Jamaica', 'Madagascar', 'Peru', 'Trinidad', 'Venezuela', 'Vietnam'], ["Reading Date", "Temperature"]]

df
```

```python
#finds all readings for our top countries after 2006 (which is the year of the first chocolate bar review in our other data set, our last review was written in 2017. Unfortunately, our weather data only goes to 2013 so we are effectively interpreting climate data from Jan 2006 to Sep 2013.)
daterange = df.loc[df["Reading Date"] <= 20130000]
daterange = daterange.loc[daterange["Reading Date"] > 19920000]
daterange
```

```python
#converting the reading dates into datetime type

pd.to_datetime(daterange["Reading Date"], format='%Y%m%d', errors='coerce')

daterange.head()
```

```python
daterange.dtypes
```

```python
daterange.groupby("Reading Date")
daterange
```

```python
#converts reading date to datetime format and separates out into years and months.
#renamed months to names and cleaned look of table

daterange['DateTime'] = daterange['Reading Date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
daterange['Year'] = daterange['DateTime'].dt.year
daterange['Month'] = daterange['DateTime'].dt.month


#monthnames = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
#daterange = daterange.replace({"Month":monthnames})

del daterange["Reading Date"]

daterange.dtypes
```

```python
#pivot table summarizes country temp. information by month, year

yearlyclimatedata =pd.pivot_table(daterange,values=['Temperature'],index=['Year'],columns=['Country'])
yearlyclimatedata.head()
```
