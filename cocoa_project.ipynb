```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv_path = "Resources/flavors_of_cacao.csv"
flavors_cacao = pd.read_csv(csv_path)
flavors_cacao.head()
```

```python
flavors_cacao.info()
```

```python
#get a list of columns 

flavors_cacao.columns
```

```python
# rename columns

renamed_flavors_cacao = flavors_cacao.rename(columns={"Company \n(Maker-if known)": "Company",
                                                      "Specific Bean Origin\nor Bar Name": "Specific Bean Origin",
                                                      "Review\nDate": "Review Date",
                                                      "Cocoa\nPercent": "Cocoa Percent",
                                                      "Company\nLocation": "Company Location",
                                                      "Bean\nType": "Bean Type",
                                                      "Broad Bean\nOrigin": "Broad Bean Origin"
                             })
renamed_flavors_cacao.columns = [c.replace('\n', ' ') for c in renamed_flavors_cacao.columns]
renamed_flavors_cacao.head()
```

```python
renamed_flavors_cacao.columns
```

```python
renamed_flavors_cacao.dtypes
```

```python
# fill space with NaN

fillna_flavors_cacao = renamed_flavors_cacao.replace(r'^\s*$', np.nan, regex=True)
fillna_flavors_cacao.head(10)
```

```python
# drop NaN values from the Broad Bean Origin column only

dropna_flavors_cacao = fillna_flavors_cacao.dropna(subset=['Broad Bean Origin'])

dropna_flavors_cacao.head(10)
```

```python


top_ten = dropna_flavors_cacao.sort_values(by='Rating', ascending=False)
top_ten.head(10)
```

```python

top_ten_df = top_ten[['Cocoa Percent', 'Company Location', 'Rating', 'Bean Type', 'Broad Bean Origin']]
top_ten_df.head(20)

# I would like to represent the who are the best bean types, best country and cocoa percent
# remove index better
```

```python
tops_df_country =pd.pivot_table(top_ten_df,values=['Rating'],index=['Broad Bean Origin'],columns=['Bean Type'])

# math_scores_by_grade.index.name=None

tops_df_country 
```

```python
bottom_ten = dropna_flavors_cacao.sort_values(by='Rating', ascending=True)

bottom_ten.head(10)
```

```python
# replace bean name

dropna_flavors_cacao['Bean Type'] = dropna_flavors_cacao['Bean Type'].replace({
    'Forastero (Nacional)': 'Forastero', 'Forastero (Arriba)': 'Forastero',
     'Criollo (Porcelana)': 'Criollo', 'Forastero (Parazinho)': 'Forastero',
     'Forastero (Arriba) ASS': 'Forastero', 'Criollo (Amarru)': 'Criollo',
     'Nacional (Arriba)': 'Nacional', 'Amazon, ICS': 'Amazon',
     'Amazon mix': 'Amazon', 'Criollo (Ocumare 61)': 'Criollo',
     'Forastero (Catongo)': 'Forastero', 'Criollo (Ocumare 77)': 'Criollo',
     'Trinitario (85% Criollo)': 'Trinitario, Criollo', 'Blend-Forastero,Criollo': 'Criollo, Forastero',
     'Criollo (Wild)': 'Criollo', 'Forastero(Arriba, CCN)': 'Forastero',
     'Trinitario (Scavina)': 'Trinitario', 'Trinitario, TCGA': 'Trinitario',
     'Criollo (Ocumare 67)': 'Criollo', 'Forastero (Arriba) ASSS': 'Forastero',
     'Criollo, +': 'Criollo', 'Forastero (Amelonado)': 'Forastero',
     'Criollo (Ocumare)': 'Criollo', 'Trinitario (Amelonado)': 'Trinitario',
    'Trinitario, Criollo': 'Criollo, Trinitario', 'Trinitario, Forastero': 'Forastero, Trinitario'
   
    
})
    
```

```python
dropna_flavors_cacao['Bean Type'].value_counts()

#the most used beans are Trinitario,Forastero, Criollo  
# find average rating for each of them
```

```python
#keep only mostly used beans

beans_cacao_df = dropna_flavors_cacao[(dropna_flavors_cacao.values  == "Trinitario")|(dropna_flavors_cacao.values  == "Forastero" ) |(dropna_flavors_cacao.values  == "Criollo" )] 
beans_cacao_df.head(10)
```

```python
bean_df = beans_cacao_df[['Cocoa Percent', 'Company Location', 'Rating', 'Bean Type', 'Broad Bean Origin']]

# bean_df.style.hide_index()

bean_df.head(10)
```
