```python
count_of_rating_by_origin = beans_cacao_df.groupby(["Broad Bean Origin"])
counts_rating_by_origin = pd.DataFrame(count_of_rating_by_origin["Rating"].count())
counts_rating_by_origin.reset_index(inplace=True)
#pick the top5 most reviews by bean origin 
counts_rating_by_origin = counts_rating_by_origin.rename(columns={"Rating":"counts_of_rating"})
rank_counts_rating_origin = counts_rating_by_origin.sort_values(by=['counts_of_rating'],ascending=False)
rank_counts_rating_origin.set_index("Broad Bean Origin", inplace=True)
most_counts_by_origin = rank_counts_rating_origin.iloc[0:10]
most_counts_by_origin.dropna()
 
```

```python
bean_origin_places = "GDP_bean_production_current$_v4-1.csv"
bean_origin_gdps = pd.read_csv(bean_origin_places)
bean_origin_gdps = bean_origin_gdps.rename(columns={"Broad Bean\nOrigin": "Broad Bean Origin"})
bean_origin_gdps = bean_origin_gdps.drop_duplicates()
bean_origin_gdps.set_index("Broad Bean Origin", inplace=True)
#bean_origin_gdps
```

```python
top_origins = bean_origin_gdps.loc[["Venezuela","Ecuador","Peru","Madagascar","Trinidad","Vietnam","Brazil","Ghana","Belize","Jamaica"]]
top_origins = top_origins.drop_duplicates()
top_origins = top_origins.drop(columns=['Income group', 'Region'])
top_origins = top_origins.apply(pd.to_numeric)
top_origins['avg_gdp'] = top_origins.mean(axis=1)
top_origins_avg_gdp = pd.DataFrame(top_origins['avg_gdp'])
top_origins_avg_gdp
```

```python
gdp_vs_most_counts = most_counts_by_origin.join(top_origins_avg_gdp,how = "outer")
gdp_vs_most_counts.dropna()
```

```python
plt.scatter(gdp_vs_most_counts['counts_of_rating'], gdp_vs_most_counts['avg_gdp'])
plt.title(f'counts of ratings vs. GDPs')
plt.xlabel('counts_of_rating')
plt.ylabel('average GDPs')
plt.grid(True)
plt.savefig("Counts_of_ratings_vs_bean_origin_GDPs.png")
plt.show()
```

```python
rating_by_origin = beans_cacao_df.groupby(["Broad Bean Origin"])
mean_rating_by_origin = pd.DataFrame(rating_by_origin["Rating"].mean())
mean_rating_by_origin.reset_index(inplace=True)
mean_rating_by_origin
rank_mean_rating_origin = mean_rating_by_origin.sort_values(by=['Rating'],ascending=False)
rank_mean_rating_origin.set_index("Broad Bean Origin", inplace=True)
top_rating_by_origin = rank_mean_rating_origin.loc[["Venezuela","Ecuador","Peru","Madagascar","Trinidad","Vietnam","Brazil","Ghana","Belize","Jamaica"]]
top_rating_by_origin
```

```python
gdp_vs_top_ratings = top_rating_by_origin.join(top_origins_avg_gdp,how = "outer")
gdp_vs_top_ratings.dropna()
```

```python
plt.scatter(gdp_vs_top_ratings['Rating'], gdp_vs_top_ratings['avg_gdp'])
plt.title(f'ratings vs. GDPs')
plt.xlabel('ratings')
plt.ylabel('average GDPs')
plt.grid(True)
plt.savefig("Ratings_vs_bean_origin_GDPs.png")
plt.show()
```

```python
count_of_rating_by_company_loc = beans_cacao_df.groupby(["Company Location"])
counts_rating_by_company_loc = pd.DataFrame(count_of_rating_by_company_loc["Rating"].count())
counts_rating_by_company_loc.reset_index(inplace=True)
#pick the top5 most reviews by company location
counts_rating_by_company_loc = counts_rating_by_company_loc.rename(columns={"Rating":"counts_of_rating"})
rank_counts_rating_companyloc = counts_rating_by_company_loc.sort_values(by=['counts_of_rating'],ascending=False)
rank_counts_rating_companyloc.set_index("Company Location", inplace=True)
most_counts_by_companyloc = rank_counts_rating_companyloc.iloc[0:10]
most_counts_by_companyloc.dropna()
```
