```python
# rename columns
flavors_cacao = flavors_cacao.rename(columns={"Company \n(Maker-if known)": "Company",
                                                      "Specific Bean Origin\nor Bar Name": "Specific Bean Origin",
                                                      "Review\nDate": "Review Date",
                                                      "Cocoa\nPercent": "Cocoa Percent",
                                                      "Company\nLocation": "Company Location",
                                                      "Bean\nType": "Bean Type",
                                                      "Broad Bean\nOrigin": "Broad Bean Origin"
                             })
flavors_cacao.columns = [c.replace('\n', ' ') for c in flavors_cacao.columns]
flavors_cacao.head()
```

```python
# fill space with NaN

flavors_cacao = flavors_cacao.replace(r'^\s*$', np.nan, regex=True)
flavors_cacao.head(10)
```

```python
flow = pd.crosstab(
    beans_cacao_df['Company Location'],
    beans_cacao_df['Review Date'],
    beans_cacao_df['Rating'], aggfunc='mean'
)
flow['tot'] = flow.sum(axis=1)
flow = flow.sort_values('tot', ascending=False)
flow = flow.drop('tot', axis=1)

fig, ax = plt.subplots(figsize=[10,6])
sns.heatmap(flow.head(20), cmap='RdBu_r', linewidths=.5)
ax.set_title('Goods Flow from Company location, mean rating by years')
```
